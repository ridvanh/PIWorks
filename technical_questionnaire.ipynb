## Question 4-5

#### If there is nothing but NaN values in a country, than its 'daily_vaccinations' mean value should be NaN aswell.

```python
df.groupby(['country']).mean()
```

#### As we can see the table above, It is only Kuwait having only NaN values. Since there is a little amount of countries that have only NaN values, we can fill them manually.

```python
df[df['country'] == 'Kuwait']
```

```python
df.loc[750, 'daily_vaccinations'] = 0
```

```python
df[df['country'] == 'Kuwait']
```

```python
min_values = df.groupby('country')['daily_vaccinations'].min()
```

```python
min_values
```

```python
df[df['daily_vaccinations'].isna()]['country'].value_counts()
```

#### Since we have only one NaN value for each country, we can match them with min values easily

```python
countries_to_fill = df[df['daily_vaccinations'].isna()]['country'].value_counts().index
```

```python
countries_to_fill
```

```python
fill_value = min_values.loc[countries_to_fill]
```

```python
fill_value
```

```python
df['daily_vaccinations'].fillna(pd.Series(fill_value.values), inplace=True)
```

```python
df['daily_vaccinations']
```

```python
medians = df.groupby('country')['daily_vaccinations'].median()
```

```python
medians
```

```python
top3_idx = medians.nlargest(3).index
top3_val = medians.nlargest(3).values

top3 = pd.DataFrame({
    'country' : top3_idx,
    'daily_vaccinations (median)' : top3_val
})
```
